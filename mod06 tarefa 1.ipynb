```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Módulo 6 tarefa 1

#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
nascidos_df = pd.read_csv('SINASC_RO_2019.csv')
nascidos_df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.close('all')
plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2,3)

sns.countplot(ax=axes[0,0], x='APGAR1', data=nascidos_df)
sns.countplot(ax=axes[0,1], x='APGAR5', data=nascidos_df)
sns.countplot(ax=axes[0,2], x='SEXO', data=nascidos_df)
g = sns.countplot(ax=axes[1,0], x='GESTACAO', data=nascidos_df)
g.tick_params(axis='x',rotation=30)
sns.countplot(ax=axes[1,1], x='ESTCIVMAE', data=nascidos_df)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5',y='CONSULTAS',data=nascidos_df)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure',figsize=(15,10))
fig, axes = plt.subplots(2,2)

sns.countplot(ax=axes[0,0], x='SEXO', data=nascidos_df)
sns.countplot(ax=axes[0,1], x='ESTCIVMAE', data=nascidos_df)
sns.countplot(ax=axes[1,0], x='CONSULTAS', data=nascidos_df)
g = sns.countplot(ax=axes[1,1], x='GESTACAO', data=nascidos_df)
g.tick_params(axis='x', rotation=30)
```
